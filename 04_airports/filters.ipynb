```python
import pandas as pd
data = {
    'COUNTRY': ['USA', 'Canada', 'USA', 'Germany', 'Canada', 'Germany', 'USA', 'USA'],
    'AIRPORT': ['JFK', 'YYZ', 'LAX', 'FRA', 'YVR', 'MUC', 'ORD', 'SFO'],
    'PASSENGERS_MILLIONS': [61, 50, 84, 69, 22, 48, 54, 57]
}
df = pd.DataFrame(data)
df
```

```python
#liczba lotnisk ogólnie
```

```python
airports_amount = df['AIRPORT'].value_counts().sum()
airports_amount
```

```python
#Liczba lotnisk w danym kraju
```

```python
airports_amount = df['AIRPORT'].value_counts()
airports_amount
```

```python
#średnia liczba pasażerów (w milionach) na lotnisko w każdym kraju
```

```python
df.groupby('COUNTRY')['PASSENGERS_MILLIONS'].mean()
```

```python
#Kraje posortowane według średniej liczby pasażerów (malejąco)
```

```python
df.groupby('COUNTRY')['PASSENGERS_MILLIONS'].mean().sort_values(ascending = False)
```
